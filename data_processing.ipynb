```python
# Prepare figure
fig = plt.figure(figsize=(15, 10))

# Obtain correlation matrix. Round the values to 2 decimal cases. Use the DataFrame corr() and round() method.
corr = np.round(data[selected_columns].corr(method="spearman"), decimals=2)

# Build annotation matrix (values above |0.6| will appear annotated in the plot)
mask_annot = np.absolute(corr.values) >= 0
annot = np.where(mask_annot, corr.values, np.full(corr.shape,"")) # Try to understand what this np.where() does

# Plot heatmap of the correlation matrix
sns.heatmap(data=corr, annot=annot, cmap=sns.diverging_palette(220, 10, as_cmap=True),
            fmt='s', vmin=-1, vmax=1, center=0, square=True, linewidths=.5)

# Layout
fig.subplots_adjust(top=0.95)
fig.suptitle("Correlation Matrix", fontsize=20)

plt.show()
```

```python
# Selected features for outlier detection
selected_columns = ['heritage', 'UNESCO', 'Nbwikis', 'TotalWords',
                    'Average Rating', 'Total number of ratings',
                    'longitude', 'latitude']

# Outlier detection using IQR method
Q1 = data[selected_columns].quantile(0.25)
Q3 = data[selected_columns].quantile(0.75)
IQR = Q3 - Q1

# Detect outliers (True for outliers)
outlier_condition = (data[selected_columns] < (Q1 - 1.5 * IQR)) | (data[selected_columns] > (Q3 + 1.5 * IQR))

# Add a new column to flag outliers in the dataset
data['is_outlier'] = outlier_condition.any(axis=1)

# Show all outliers
outliers = data[data['is_outlier']]
print(outliers[['placeLabel', 'heritage', 'UNESCO', 'Nbwikis', 'TotalWords',
                'Average Rating', 'Total number of ratings', 'longitude', 'latitude']])

# Plot outliers using a boxplot for each numerical feature
plt.figure(figsize=(12,8))
for i, column in enumerate(selected_columns):
    plt.subplot(2, 4, i+1)
    sns.boxplot(x=data['is_outlier'], y=data[column])
    plt.title(f'Outlier Detection in {column}')
plt.tight_layout()
plt.show()

```

```python
# Remove outliers from the original dataset
cleaned_data = data[~data['is_outlier']]

# Display the shape of the original and cleaned dataset
print(f"Original dataset shape: {data.shape}")
print(f"Cleaned dataset shape: {cleaned_data.shape}")

# Optionally, reset the index of the cleaned dataset
cleaned_data.reset_index(drop=True, inplace=True)

# Show the first few rows of the cleaned dataset
print(cleaned_data.head())
```

```python
clustering_features_new = [ 'LgWrdContent', 'NbPhotos', 'TotalWords',
                            'Average Rating', 'Total number of ratings',
                            'heritage', 'UNESCO']
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Range of clusters to try
cluster_range = range(1, 11)
inertia = []

# Compute KMeans for each number of clusters
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(clustering_features_new )
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(cluster_range)
plt.grid()
plt.show()
```
