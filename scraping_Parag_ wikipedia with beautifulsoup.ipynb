```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def initialize_dataframe():
    # Initialise la DataFrame avec les colonnes nécessaires
    df = pd.DataFrame(columns=["Nom du site", "Titres", "Paragraphes"])
    return df

def extract_wikipedia_content(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Vérifie les erreurs HTTP

        soup = BeautifulSoup(response.content, 'html.parser')

        # Trouver tous les sous-titres (balises <h2> et <h3>)
        subtitles = soup.find_all(['h2', 'h3'])

        # Créer une liste pour stocker les données
        data = []

        # Parcourir les sous-titres pour extraire les paragraphes associés
        for subtitle in subtitles:
            title_text = subtitle.get_text().strip()
            paragraphs = []
            # Trouver tous les paragraphes suivants jusqu'au prochain sous-titre
            for sibling in subtitle.find_next_siblings():
                if sibling.name in ['h2', 'h3']:
                    break
                if sibling.name == 'p':
                    paragraph_text = sibling.get_text().strip()
                    if paragraph_text:
                        paragraphs.append(paragraph_text)

            if paragraphs:
                # Ajouter les données à la liste
                data.append({"Nom du site": url, "Titres": title_text, "Paragraphes": " ".join(paragraphs)})

        return data

    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error: {e}")
        return []  # Retourner une liste vide en cas d'erreur HTTP
    except requests.exceptions.RequestException as e:
        print(f"Request Error: {e}")
        return []  # Retourner une liste vide en cas d'erreur de requête

# Liste de liens à traiter
links = [
    "https://fr.wikipedia.org/wiki/Taj_Mahal",
    "https://fr.wikipedia.org/wiki/Amphith%C3%A9%C3%A2tre_de_Carthage"

]

# Initialise une DataFrame vide ou chargement du fichier CSV existant
existing_file = "wikipedia_content_total2.csv"
try:
    df_total = pd.read_csv(existing_file)
except FileNotFoundError:
    df_total = initialize_dataframe()

# Pour chaque lien, extraire les informations et ajouter à df_total
for link in links:
    # Extraire les données pour le lien actuel
    data = extract_wikipedia_content(link)

    # Ajouter les données extraites à la DataFrame existante
    df = pd.DataFrame(data)
    df["Nom du site"] = link  # Ajouter la colonne avec le lien d'origine
    df_total = pd.concat([df_total, df])

# Réinitialiser les index de la DataFrame combinée
df_total.reset_index(drop=True, inplace=True)

# Sauvegarder la DataFrame mise à jour dans le même fichier CSV
df_total.to_csv(existing_file, index=False)

# Afficher la DataFrame résultante
print(df_total)
```
