```python
import pandas as pd
import urllib.parse

def get_google_maps_url(place_name):
    # Encode the place name to handle special characters
    formatted_place_name = urllib.parse.quote(place_name)

    # Construct the Google Maps URL
    google_maps_url = f"https://www.google.com/maps/search/?api=1&query={formatted_place_name}"

    return google_maps_url

# List of places
places = [
    "Eglise Sainte-Felicite et Sainte-Perpétue de Tebourba",
    "Place de la Concorde",
    "Tour Eiffel"
]

# Initialize an empty list to store data
data = []

# Loop through each place, get Google Maps URL and append to data list
for place in places:
    url = get_google_maps_url(place)
    data.append({'Place Name': place, 'Google Maps URL': url})

# Create a DataFrame from the data list
df = pd.DataFrame(data)



# Export the DataFrame to CSV
df.to_csv('places.csv', index=False, encoding='utf-8')
print("\nLes données ont été écrites dans places.csv.")
```
